**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

```python
properties.dropna(axis = 1, how = 'all', inplace = True)
properties2=properties.transpose()
new_header=properties2.iloc[0]
properties3=properties2.iloc[1:]
properties4=properties3
properties4.columns=new_header
properties5=properties4.reset_index()
properties5
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
#set the appropriate column name for the postal code
properties6=properties5.rename(columns={'index': 'borough', pd.NaT: 'post_code'})
#drop the columns with 'NaN' in all rows
properties6.dropna(axis = 0, how = 'all', inplace = True)
## selecting only London Boroughs. I did a search on Wikipedia to look at the London boroughs
properties7=properties6.iloc[1:33]
properties7
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
#Here I have two tables:
#1) the original table with 'borough','post_code','date','Year','Month','average_housing_price' columns -> properties9
#2) a new table with the average housing prices for each year for each borough -> avg_hous_price_year

#melt dataframe to reduce number of columns
properties8=properties7.melt(id_vars=['borough','post_code'], var_name='date',value_name='average_housing_price')
properties8

#I'd like to add year and month columns to the dataframe
properties9=properties8
properties9['Month']=properties9['date'].dt.month
properties9['Year']=properties9['date'].dt.year
#I am moving the prices to last
properties9=properties9[['borough','post_code','date','Year','Month','average_housing_price']]

#I am creating a new table with the average housing prices for each year for each borough
avg_hous_price_year=properties9.sort_values(by=['borough','date']).groupby(['borough','Year'])['average_housing_price'].mean().reset_index(name='Average Housing Price by Year')

properties9
avg_hous_price_year
```

```python
#Here I am adding one column for the difference in the average housing prices in 2021 compared to 1995. The second column
#is the ratio of the value in 2021 over that in 1995. Then, I plot these in barplots using seaborn

year1=1995
year2=2021
avg_hous_price_year1=avg_hous_price_year[avg_hous_price_year['Year']==year1]
avg_hous_price_year2=avg_hous_price_year[avg_hous_price_year['Year']==year2]

avg_hous_price_year2=avg_hous_price_year2.drop(['Year'],axis=1).reset_index(drop=True).rename(columns={'Average Housing Price by Year':'%s Average Housing Price by Year' %(year2)})
avg_hous_price_year1=avg_hous_price_year1.drop(['Year'],axis=1).reset_index(drop=True).rename(columns={'Average Housing Price by Year':'%s Average Housing Price by Year' %(year1)})
ratio_hous_price_year1_year2=avg_hous_price_year1.join(avg_hous_price_year2['%s Average Housing Price by Year' %(year2)])


ratio_hous_price_year1_year2['%s/%s Ratio' %(year2,year1)]=ratio_hous_price_year1_year2['%s Average Housing Price by Year' %(year2)]/ratio_hous_price_year1_year2['%s Average Housing Price by Year' %(year1)]
ratio_hous_price_year1_year2['%s-%s Difference' %(year2,year1)]=ratio_hous_price_year1_year2['%s Average Housing Price by Year' %(year2)]-ratio_hous_price_year1_year2['%s Average Housing Price by Year' %(year1)]
ratio_hous_price_year1_year2['%s-%s Percent Increase' %(year2,year1)]=((ratio_hous_price_year1_year2['%s Average Housing Price by Year' %(year2)]-ratio_hous_price_year1_year2['%s Average Housing Price by Year' %(year1)])/ratio_hous_price_year1_year2['%s Average Housing Price by Year' %(year1)])*100
ratio_hous_price_year1_year2.sort_values(by='%s/%s Ratio' %(year2,year1),ascending=False)

```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:

```python
#here, I'd also like to add to the function the choice of the year. For me it is also more intuitive to divide the value
#for the latest year (year2) over the value of the older year (year1)
def create_price_ratio(filtered_df,year1,year2):
    ''' input: a filter of df, specifically where this filter constrains the London_Borough, as an argument, and year1 and year2
        to be used in the function
        what it does: Get the Average Price for that Borough, for year1 and year2.
                      Calculate the ratio of the Average Price for year2 divided by the Average Price for year1
        output: the ratio'''

    new_df=filtered_df.groupby('Year')['average_housing_price'].mean().reset_index(name='Average Housing Price by Year')
    price_year1=new_df[new_df['Year']==year1]['Average Housing Price by Year'].item()
    price_year2=new_df[new_df['Year']==year2]['Average Housing Price by Year'].item()
    ratio=price_year2/price_year1

    return ratio
```
